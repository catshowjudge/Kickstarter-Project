# Kickstarter Dataset: Exploration

#### This notebook is the first part of my exploration into the Kickstarter dataset. 

This dataset has information for more than 300k Kickstarter projects launched between 2009 and 2018, including their titles, descriptions, project categories, geographical location, launch dates and deadlines, target funding goal, final total funding, outcomes, and number of contributors. This information can allow us to look for investment patterns across time, region, and project category, as well as the success or failure of various kinds of projects, defined as whether or not the funding goal was met. The dataset is available in the open source at: https://www.kaggle.com/ukveteran/kickstarter

## Imports 
Importing the relevant libraries for creating dataframes and visualizing data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
from scipy import stats
from numpy.random import seed
from scipy.stats import t
%matplotlib inline
print('Done!')
```

## Reading in the CSV file

```python
df = pd.read_csv('kickstarter-projects.csv')
df.head()
```

## Data Cleaning

### Ensuring that the "launched" and "deadline" columns are in datetime format
Converting them to datetime64 so that I can extract year, month, and day data more easily later in the notebook

```python
df["launched"] = df["launched"].astype("datetime64")
df["deadline"] = df["deadline"].astype("datetime64")
```

### Adding a "duration" column to show the length of each project in days
By subtracting the deadline from the launch date, we can see the duration of crowdfunding for each project. This difference is saved as an integer in a newly created column.

```python
df['duration'] = (df['deadline'] - df['launched']).dt.days

#Let's take a look at the new column
df.head()
```

### Getting rid of 1970 projects rows
This seems like a mistake in the data, so in order to get the cleanest data possible, these rows will be excluded

```python
df1 = df[df['launched'].dt.year != 1970]
len(df1)
```

### Removing the rows with N,0" as the country, since this is unknown
Since we are unsure what this is, we should take it out to keep the data cleaner

```python
df2 = df1[df1['country'] != 'N,0"']
len(df2)
```

### Removing rows with "live" and "undefined" values from the Status column

Projects fall into one of six status categories in this dataset: successful, failed, live, suspended, cancelled, undefined. 
- Successful implies that the project has met its funding goal within the given duration.
- Failed means that the project has failed to achieve its goal within the time frame.
- Live means that the project was ongoing at the time of data collection and had not arrived at the deadline
- Suspended means that the project has been put on hold
- Cancelled means that the project was cut off before the deadline
- undefined means that it has an unknown status

Ideally we would like to define projects as either successful or failed.
- Live and undefined projects can not be determined to have either failed or succeeded, so we will get rid of these

```python
a = df2[(df2.state != 'undefined') & (df2.state != 'live')]
len(a)
```

```python
a.state.value_counts()
```

### Changing "canceled" and "suspended" values to "failed" in the Status column
Canceled and suspended projects have technically failed to meet their funding goals within the specified time frame, so we will reclass these as failed

```python
a.replace(to_replace=['canceled', 'suspended'],
           value='failed', 
           inplace=True)
a.state.value_counts()
```

### Changing "successful" and "failed" in the "state" column to 1 and 0 in order to calculate the rate of success
In order to examine the columns with categorical data according to whether they have succeeded or failed, we will first convert 'successful' and 'failed' to numbers (1 and 0, respectively).
- By taking the mean of the status column for each grouping within a categorical column, we can see the rate of success

```python
# a.replace(to_replace=['successful', 'failed'],
#            value=[1, 0], 
#            inplace=True)
#a.state.value_counts()
```

```python
equiv = {'successful':1, 'failed':0}
a['rate'] = a['state'].map(equiv)
a.head(10)
```

#### Statistical tests for the goal column

T-Test

```python
#inspect the means and standard deviations for each column, separated by success or failure
m1 = a['goal'].loc[a.rate == 1].mean()
m0 = a['goal'].loc[a.rate == 0].mean()
sd1 = a['goal'].loc[a.rate == 1].std()
sd0 = a['goal'].loc[a.rate == 0].std()
print(m1, m0, sd1, sd0)
```

```python
#two-sided t-test for two independent samples, not assuming equal variance
''' 
This is a two-sided test for the null hypothesis that 2 independent samples 
have identical average (expected) values.
''' 

success_goal = a['goal'].loc[a.rate == 1]
failed_goal = a['goal'].loc[a.rate == 0]
stats.ttest_ind(success_goal, failed_goal, equal_var=False)
```

Bootstrap replicates for differences in standard deviation and mean

- resamples the divided data
- includes plots of differences in SD and mean

```python
#set a function for replication
#Our data set is large so we will use 50% only
#This only needs to be set once

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates function from Datacamp statistical thinking with python 2."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = func(np.random.choice(data, size=round(len(data)/2)))

    return bs_replicates
```

```python
#computing the differences in standard deviations over 1000 random samples
np.random.seed(47)
bs_diff = draw_bs_reps(success_goal, np.std, size=1000) - draw_bs_reps(failed_goal, np.std, size=1000)

#calculating the 95% confidence interval
CI = np.percentile(bs_diff, [2.5, 97.5])
print(CI)
```

```python
# Plot of difference in standard deviations


#set more variables for plotting
bs_mean = np.mean(bs_diff)
bs_34right = np.mean(bs_diff) + np.std(bs_diff)
bs_34left = np.mean(bs_diff) - np.std(bs_diff)

#plot histogram of differences of standard deviation
_ = plt.hist(bs_diff, bins=30)
_ = plt.xlabel('Difference of Std Deviation between successful and failed projects')
_ = plt.ylabel('number of samples')
_ = plt.title('Distribution of differences in Standard Deviation between successful and failed projects')
_ = plt.axvline(bs_mean, color='r')
_ = plt.axvline(bs_34right, color='r', linestyle='--')
_ = plt.axvline(bs_34left, color='r', linestyle='--')
_ = plt.axvline(-1697248.41, color='r', linestyle='-.')
_ = plt.axvline(-1171886.91, color='r', linestyle='-.')
```

```python
#computing the differences in means over 1000 random samples
np.random.seed(47)
bs_diff = draw_bs_reps(success_goal, np.mean, size=1000) - draw_bs_reps(failed_goal, np.mean, size=1000)

#calculating the 95% confidence interval
CI = np.percentile(bs_diff, [2.5, 97.5])
print(CI)
```

```python
# Plot of difference in mean


#set more variables for plotting
bs_mean = np.mean(bs_diff)
bs_34right = np.mean(bs_diff) + np.std(bs_diff)
bs_34left = np.mean(bs_diff) - np.std(bs_diff)

#plot histogram of differences of mean
_ = plt.hist(bs_diff, bins=30)
_ = plt.xlabel('Difference of Mean between successful and failed projects')
_ = plt.ylabel('number of samples')
_ = plt.title('Distribution of differences in Mean between successful and failed projects')
_ = plt.axvline(bs_mean, color='r')
_ = plt.axvline(bs_34right, color='r', linestyle='--')
_ = plt.axvline(bs_34left, color='r', linestyle='--')
_ = plt.axvline(-69830.46, color='r', linestyle='-.')
_ = plt.axvline(-52706.29, color='r', linestyle='-.')
```

Permutation Test

- To check if our results could have been achieved randomly

```python
#set functions for permutation testing
#we only need to set these once


#function for difference of means
def diff_of_means(data_1, data_2):
    """Difference in means of two arrays, function from Datacamp statistical thinking with python 2.."""

    # The difference of means of data_1, data_2: diff
    empirical_diff_means = np.mean(data_1) - np.mean(data_2)

    return empirical_diff_means


#function for recombining and sampling randomly from the data
def permutation_sample(data_1, data_2):
    """Generate a permutation sample from two data sets, function from Datacamp statistical thinking with python 2."""

    # Concatenate the data sets: data
    data = np.concatenate((data_1, data_2))

    # Permute the concatenated array: permuted_data
    permuted_data = np.random.permutation(data)

    # Split the permuted array into two: perm_sample_1, perm_sample_2
    perm_sample_1 = permuted_data[:len(data_1)]
    perm_sample_2 = permuted_data[len(data_1):]

    return perm_sample_1, perm_sample_2



#function for generating new arrays

def draw_perm_reps(data_1, data_2, func, size=1):
    """Generate multiple permutation replicates, function from Datacamp statistical thinking with python 2."""

    # Initialize array of replicates: perm_replicates
    perm_replicates = np.empty(size)

    for i in range(size):
        # Generate permutation sample
        perm_sample_1, perm_sample_2 = permutation_sample(data_1, data_2)

        # Compute the test statistic
        perm_replicates[i] = func(perm_sample_1, perm_sample_2)

    return perm_replicates
```

```python
# permutation test and p-value calculation using the above three functions

#random seed so it can be replicated
np.random.seed(47)
# Draw 1000 permutation replicates: perm_replicates
perm_replicates = draw_perm_reps(success_goal, failed_goal,
                                 diff_of_means, size=1000)

#calculating the empirical difference of means
empirical_diff_means = diff_of_means(success_goal, failed_goal)

# Compute p-value: p
p = np.sum(perm_replicates <= empirical_diff_means) / len(perm_replicates)

# Print the result
print('p-value =', p)
```

```python

#plot the histogram of the differences in the means from the permutation test, with the observed mean marked
_ = plt.hist(perm_replicates, bins=30)
_ = plt.xlabel('difference in means')
_ = plt.ylabel('number of samples')
_ = plt.title('Distribution of differences in means')
_ = plt.axvline(perm_replicates.mean(), color='r')
_ = plt.axvline(empirical_diff_means, color='r', linestyle='--')
```
